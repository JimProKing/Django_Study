```python
from urllib.request import urlopen,Request
from urllib.parse import urlencode

url = "http://127.0.0.1:8000/"
data = {
    'name' : '이영찬',
    'email' : 'qkq010177@gmail.com',
    'url' : 'www.google.co.kr',
}

encData = urlencode(data)
postData = bytes(encData, encoding='utf-8')
req = Request(url,data=postData)
req.add_header('Context-Type','application/x-www-form-urlencoded')
f = urlopen(req)

print(f.headers)
print(f.read().decode('utf-8'))
```

```python
from urllib.request import Request,HTTPCookieProcessor, build_opener
url = "http://127.0.0.1:8000/"
url += "cookie/"


cookie_handler = HTTPCookieProcessor()
## HTTPCookieProcessor :: 
## <urllib.request.HTTPCookieProcessor object at 0x115ace7c0>
opener = build_opener(cookie_handler)
## opener ::
## <urllib.request.OpenerDirector object at 0x1153528e0>

req = Request(url)
resp = opener.open(req)

# first response(Get) with cookie handler
print("<first Response after Get Request> \n")
print(resp.headers)
print(resp.read().decode('utf-8'))

# second request (Post) with cookie handler
print("-----------------")
data = "language = python&framework = django"
encData = bytes(data,encoding='utf-8')

req = Request(url)
resp = opener.open(req)

print("<second Response after Get Request> \n")
print(resp.headers)
print(resp.read().decode('utf-8'))
```
